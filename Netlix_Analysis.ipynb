#**Checking Null Values**

```python
df.isnull().sum()
```

#**Filling Null Values With Different Object ,As We Does Not Have Accurate Data .**

```python
df['director'].fillna('No Data Available', inplace=True)
df['cast'].fillna('No Data Available', inplace=True)
df['country'].fillna('No Data Available', inplace=True)
```

```python
df.isnull().sum()
```

#**Taking Mode**

```python
mode_rating = df['rating'].mode()[0]
mode_rating
```

#**Replacing Null Value With Mode**

```python
df['rating'].fillna(mode_rating, inplace=True)
```

```python
df.isnull().sum()
```

#**Dropping Rest Of The Null Values , As They are In Very Less Amount .**

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

#**Data Visulization**

# **Q1.What is the overall distribution of content types (movies vs. TV shows) ?**

```python
content_type_counts = df['type'].value_counts()
content_type_counts
```

```python
plt.figure(figsize=(10, 6))
colors = ['steelblue', 'salmon']
bar_plot = content_type_counts.plot(kind='bar', color=colors)
bar_plot.set_xlabel("Content Type")
bar_plot.set_ylabel("Count")
bar_plot.set_title("Distribution of Content Types")
for i, v in enumerate(content_type_counts):
    bar_plot.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

#**Q2.Which rating has the highest among all in this dataset ?**

```python
rating_counts = df.groupby('rating')['show_id'].count()
rating_counts
```

```python
plt.figure(figsize=(15, 6))
colors = plt.cm.Set3(np.linspace(0, 1, 14))
bar_plot = rating_counts.plot(kind='bar', color=colors)
bar_plot.set_xlabel("Rating")
bar_plot.set_ylabel("Count")
bar_plot.set_title("Count of Ratings")
for i, v in enumerate(rating_counts):
    bar_plot.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

# **Here we can see that rating "TV_-MA" has highest rating .**

#**Q3.Which are the top 10 countries with highest number of movies  ?**

```python
top_countries_movies = df[df['type'] == 'Movie']['country'].value_counts().head(10)
top_countries_movies
```

#**Here Dropping Index[2] Value , As It IS Replaced Value Of Null Value .**

```python
top_countries_movies = top_countries_movies.drop(top_countries_movies.index[2])
```

```python
top_countries_movies
```

```python
plt.figure(figsize=(15, 6))
ax = sns.barplot(x=top_countries_movies.values, y=top_countries_movies.index, palette='viridis')

for i, v in enumerate(top_countries_movies.values):
    ax.text(v + 10, i, str(v), color='black', fontweight='bold')

plt.xlabel('Movie Count')
plt.ylabel('Country')
plt.title('Top 10 Countries with Highest Number of Movies')

plt.show()
```

#**Q4.Analysis of Movies and Tv_Shows Year Wise  ?**

```python
yearly_counts = df.groupby(['release_year', 'type'])['show_id'].count().unstack().fillna(0)
yearly_counts
```

```python
yearly_counts.plot(kind='bar', figsize=(15, 6))
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Number of Movies and TV Shows Released Year-Wise')
plt.show()
```

#**Q5.Which are the Distribution of Movies and TV_Shows   ?**

```python
type_counts = yearly_counts.sum()
type_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.axis('equal')
plt.title('Distribution of Movies and TV Shows by Year')
plt.show()
```

#**Q6. What are the Top 10 Genres with the Highest Number of Unique Titles  ?**

```python
Genre=df.groupby(['listed_in']).agg({"title":"nunique"})
Genre
```

```python
top_10_genre = Genre.sort_values(by='title', ascending=False).head(10)
top_10_genre
```

```python
plt.figure(figsize=(17, 6))
top_10_genre.plot.barh(color='C0')
plt.xlabel('Number of Unique Titles')
plt.ylabel('Genre')
plt.title('Top 10 Genres with the Highest Number of Unique Titles')
for index, value in enumerate(top_10_genre['title']):
    plt.text(value, index, str(value))
plt.show()
```

#**Q7. Top 10 directors with most movies  ?**

```python
top_directors = df.groupby('director')['show_id'].count().sort_values(ascending=False).head(10)
top_directors
```

```python
top_directors = top_directors.drop('No Data Available')
```

```python
top_directors
```

```python
top_directors_df = pd.DataFrame({'Director': top_directors.index, 'Movie Count': top_directors.values})
```

```python
top_directors_df
```

```python
plt.figure(figsize=(13, 6))
sns.barplot(x='Movie Count', y='Director', data=top_directors_df, palette='muted')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.title('Top 10 Directors with the Most Movies')
plt.show()
```
